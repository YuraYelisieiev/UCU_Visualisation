```python
import pandas as pd
import altair as alt
```

```python
country_list = ["US", "Brazil", "India", "Russia", "Chine", "France", "Italy", "Spain", "Ukraine", "Germany", "Canada"]
df = pd.read_csv("covid19_global_confirmed_cases.csv", parse_dates=['date'])
df = df[df['country_region'].isin(country_list)]
```

```python
df.head()
```

```python
alt.Chart(df).mark_area().encode(
    x = alt.X(field="date", type="temporal"),
    y = alt.Y(field="confirmed_cases", type='quantitative', stack=True),
    color = alt.Color(field = 'country_region', type = 'nominal')
).configure_view(width = 1500, height = 500)
```

From this plot we can obtain information about the amount of confirmed COVID-19 cases and compare them to other countries. We can conclude that the biggest amount has Brazil with a huge gap towards Canada, France, Germany and India.
